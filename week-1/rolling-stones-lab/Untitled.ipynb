```python
def find_by_name(name):
    for album_dict in list_of_album_dicts:
        if album_dict['album'] == name:
            return album_dict
    return None
    
def find_by_rank(number):
    for album_dict in list_of_album_dicts:
        if int(album_dict['number']) == number:
            return album_dict
    return None 
    
def find_by_year(year):
    albums_by_year = []
    for album_dict in list_of_album_dicts:
        if int(album_dict['year']) == year:
            albums_by_year.append(album_dict)        
    return albums_by_year

def find_by_years(year1 , year2):
    albums_by_year = []
    for album_dict in list_of_album_dicts:
        if year1 <= int(album_dict['year']) <= year2:
            albums_by_year.append(album_dict)        
    return albums_by_year

def find_by_rank(rank):
    for album_dict in list_of_album_dicts:
        if int(album_dict['number']) == rank:
            return album_dict

def find_by_ranks(rank1, rank2):
    albums_by_rank = []
    for album_dict in list_of_album_dicts:
        if int(rank1) <= int(album_dict['number']) <= int(rank2):
            albums_by_rank.append(album_dict)        
    return albums_by_rank

# def find_by_year(year):
#     return list(filter(lambda album_dict: album_dict['year'] == year, list_of_album_dicts))
    
# def find_by_years(year1, year2):
#     return list(filter(lambda album_dict: year1 <= album_dict['year'] <= year2, list_of_album_dicts))

# def find_by_ranks(rank1, rank2):
#     return list(filter(lambda album_dict: rank1 <= album_dict['number'] <= rank2, list_of_album_dicts))

def all_titles(list_of_album_dicts):
    titles = []
    for album_dict in list_of_album_dicts:
        title = album_dict['album']
        titles.append(title)
    return titles

def all_artists(list_of_album_dicts):
    artists = []
    for album_dict in list_of_album_dicts:
        artist = album_dict['artist']
        artists.append(artist)
    return artists
        
```

```python
def most_albums(list_of_album_dicts):
    artists = all_artists(list_of_album_dicts)
    return _get_max_freq(artists) 
```

```python
def _get_max_freq(elements):
    freq = {}
    for element in elements:
        if element not in freq:
            freq[element] = 1
        else:
            freq[element] += 1
    return max(freq.items(),key=lambda t:t[1])[0]
```

```python
import string

def most_words(list_of_album_dicts):
    titles = all_artists(list_of_album_dicts)
    def filter_out_special_chars(string_):
        good_char = string.ascii_letters + ' '
        good_list = filter(lambda c : c in good_char, string_)
        good_str = ''.join(good_list)
        return good_str
    def get_words(string_):
        good_str = filter_out_special_chars(string_)
        list_of_words = good_str.split(' ')
        return list_of_words
    words_in_titles = [get_words(title) for title in titles]
   
    words = sum(words_in_titles,[])
    
    return _get_max_freq(words)
        
most_words(list_of_album_dicts)    
    
```

```python
def list_of_years():
    list_yrs = []
    for album_dict in list_of_album_dicts:
        list_yr = album_dict['year']
        list_yr_int = int(list_yr)
        list_yrs.append(list_yr_int)
    return list_yrs

import matplotlib.pyplot as plt
%matplotlib inline
def make_hist_albums_by_decade():
    plt.clf
    return plt.hist(list_of_years(), bins=[1950,1960,1970,1980,1990,2000,2010,2020])
make_hist_albums_by_decade()
```

```python
def list_of_genres():
    list_genres = []
    for album_dict in list_of_album_dicts:
        genre = album_dict['genre'].split(',')
        genre = [s.strip() for s in genre]
        list_genres.append(genre)
    return sum(list_genres, [])

def make_hist_albums_by_genre():
    plt.clf()
    plt.hist(list_of_genres())
    plt.show()

list_of_genres()
make_hist_albums_by_genre()
```
